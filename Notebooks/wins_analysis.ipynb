```python
import pandas as pd
%matplotlib inline
```

```python
pdb = pd.read_csv('/Users/Alex/Coding/weather_db/Data/prediction_db_analysis.csv')
```

```python
pdb.groupby('province').mean()[['ML_win', 'TWN_win', 'EC_win']].plot(figsize=(15,10))
```

```python
pdb.groupby('province').mean()[['ML_win', 'TWN_win', 'EC_win']].plot(figsize=(20,10), kind='box')
```

```python
pdb.groupby('date').mean()[['ML_win', 'TWN_win', 'EC_win']].plot(figsize=(20,10))
```

```python
pdb.groupby('date').mean()[['ML_win', 'TWN_win', 'EC_win']].plot(figsize=(20,10), kind='box')
```

```python
pdb.groupby('region').mean()[['ML_win', 'TWN_win', 'EC_win','longitude']].reset_index().sort_values('longitude').plot(figsize=(20,10), kind='line', x='region', y=['ML_win', 'TWN_win', 'EC_win'])
```

```python
pdb.groupby('region').mean()[['ML_win', 'TWN_win', 'EC_win','longitude']].reset_index().sort_values('longitude').plot(figsize=(20,10), kind='bar', x='region', y=['ML_win', 'TWN_win', 'EC_win'])
```

```python
pdb.groupby('province').mean()[['ML_win', 'TWN_win', 'EC_win','longitude']].reset_index().sort_values('longitude').plot(figsize=(20,10), kind='bar', x='province', y=['ML_win', 'TWN_win', 'EC_win'])
```
